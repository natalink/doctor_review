### Making one dataframe out of two large excel files

```python
import pandas as pd
import re

all_data = []
df1 = pd.read_excel('all_data_1.xlsx', sheet_name="Sheet1")
df2 = pd.read_excel('all_data_2.xlsx', sheet_name="Sheet1")
all_data.append(df1) 
all_data.append(df2)
df = pd.concat(all_data, ignore_index=True)
```

### Processing columns
Several columns are selected, the values are processed by regular expressions (leaving just numbers without text)

```python
df[df.columns[8]] = df[df.columns[8]].str.extract(r'(\d+)$', expand=True) 
# regex for another value in the above column G: df[df.columns[8]] = df[df.columns[8]].str.extract(r'(\d+)\/')
#df['rating(post)'] = df['rating(post)'].str.extract(r'(^\d)') # Many undefined values, ignore this column
df['no. of letters'] = df['no. of letters'].str.extract(r'\((\d+)\)', expand=True) # If repeated twice, for some reasons all values are turned into Nans :((())
df['no. of articles'] = df['no. of articles'].str.extract(r'^(\d+)', expand=True)
df['no. of patients'] = df['no. of patients'].str.extract(r'^(\d+)', expand=True)
#print (df)
cleaned_dataframe = df[['rating(recom)', df.columns[8], 'no. of letters', 'no. of articles', 'no. of patients']]
print (cleaned_dataframe)

drop_nans_dataframe = cleaned_dataframe.dropna(how='any')
print (drop_nans_dataframe)
```
